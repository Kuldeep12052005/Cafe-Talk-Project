```python
import pandas as pd
from colorama import Fore, Back, Style

class CafeTalk:
    def Cafe(self):
        print("\t\t\t CafeTalk")
        print("\t\t\t Since-2000")
        print("\t\t\t www.CafeTalk.com")
        print("\t\t\t MENU")

    def food(self):
        a = ["Indian", "Korean", "Japanese", "Chinese"]
        for item in a:
            print("\n", item)

    def Chinese(self):
        a = ["Hot Pot", "Mapo Tofu", "Char Siu", "Chicken Fried Rice", "Wonton Soup", "Cantonese Spring Rolls", "Chow Mein"]
        b = [250, 120, 400, 300, 800, 90, 350]
        chinese = {"Chinese food types": a, "Price": b}
        df = pd.DataFrame(chinese)
        print("\nChinese foods\n")
        print(df)

        z = []
        while True:
            posi = int(input("Enter the position of the food item: "))
            q = int(input("Enter the quantity: "))
            price = chinese["Price"][posi]
            total_price = price * q
            z.append(total_price)

            a = input("Do you want to reorder? (yes/no): ")
            if a == "no":
                print("Thank you for your order!")
                break

        l = sum(z)
        m = l * 0.06
        k = l + m
        print("Service tax = 6%")
        print("Total amount:", l, "Rupees")
        print("Total amount including service tax:", k, "Rupees")

    def Korean(self):
        a = ["Kimchi", "Bibimbap", "Bulgogi", "Jajangmyeon", "Samgyeopsal", "Korean Fried Chicken", "Red Rice Cakes","Tteokbokki"]
        b = [200, 100, 350, 400, 100, 345, 250]
        korean = {"Korean food types": a, "Price": b}
        df = pd.DataFrame(korean)
        print("\nKorean foods\n")
        print(df)

        z = []
        while True:
            po = int(input("Enter the position of the food item:"))
            q = int(input("Enter the quantity: "))
            price = korean["Price"][po]
            total_price = price * q
            z.append(total_price)

            a = input("Do you want to reorder? (yes/no): ")
            if a == "no":
                print("Thank you for your order!")
                break

        l = sum(z)
        m = l * 0.06
        k = l + m
        print("Service tax = 6%")
        print("Total amount:", l, "Rupees")
        print("Total amount including service tax:", k, "Rupees")

    def Japanese(self):
        a = ["Sushi", "Ramen", "Tempura", "Okonomiyaki", "Japanese Curry Rice", "Miso Soup", "Yakitori", "Onigiri"]
        b = [250, 300, 200, 100, 500, 700, 250, 150]
        japanese = {"Japanese food types": a, "Price": b}
        df = pd.DataFrame(japanese)
        print("\nJapanese foods\n")
        print(df)

        z = []
        while True:
            po = int(input("Enter the position of the food item (0-7): "))
            q = int(input("Enter the quantity: "))
            price = japanese["Price"][po]
            total_price = price * q
            z.append(total_price)

            a = input("Do you want to reorder? (yes/no): ")
            if a == "no":
                print("Thank you for your order!")
                break

        l = sum(z)
        m = l * 0.06
        k = l + m
        print("Service tax = 6%")
        print("Total amount:", l, "Rupees")
        print("Total amount including service tax:", k, "Rupees")

    def Indian(self):
        a = ["Misal Pav", "Kosha Mangsho", "Makki Di Roti & Sarso Da Saag", "Dhokla", "Rogan Josh", "Papaya Khar", "Litti Chowkha",  "Biryani", "Fish Curry"]
        b = [80, 200, 350, 120, 200, 150, 250, 400, 500]
        indian = {"Indian food types": a, "Price": b}
        df = pd.DataFrame(indian)
        print("\nIndian foods\n")
        print(df)

        z = []
        while True:
            po = int(input("Enter the position of the food item (0-8): "))
            q = int(input("Enter the quantity: "))
            price = indian["Price"][po]
            total_price = price * q
            z.append(total_price)

            a = input("Do you want to reorder? (yes/no): ")
            if a == "no":
                print("Thank you for your order!")
                break

        l = sum(z)
        m = l * 0.06
        k = l + m
        print(Fore.PINK + Back.RED + Style.BRIGHT + "Service tax = 6%")
        print("Total amount:", l, "Rupees")
        print("Total amount including service tax:", k, "Rupees")


obj = CafeTalk()
obj.Cafe()
obj.food()

a = input("Enter the Type of food")

if a == "Indian":
    obj.Indian()
elif a == "Chinese":
    obj.Chinese()
elif a == "Korean":
    obj.Korean()
elif a == "Japanese":
    obj.Japanese()
else:
    print("Wrong option. Please choose the right option.")
```
